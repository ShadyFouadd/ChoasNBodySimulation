> - Initial Conditions: Plummer Sphere
> <ol>
> <li> density profile </li>
> <center> $\rho(r) = \frac{3M_0}{4\pi a^3} \left(1 + \frac{r^2}{a^2}\right)^{-\frac{5}{2}}$ </center>
> <li> Mass enclosed within radius $r$</li>
> <center> $M(<r) = M_0 \frac{r^3}{\left(r^2 + a^2\right)^{3/2}} $ </center>
> <center> $r(m) = a \left(m^{-2/3} - 1 \right)^{-\frac{1}{2}}$, where $m = \frac{M}{M_0}$ </center>
> </ol>

```python
import numpy as np
import matplotlib.pylab as plt

from matplotlib import rc
rc('animation', html='html5')
rc("font", weight="bold"); rc("font", size="20")
rc("axes"); rc("axes",labelsize="20"); rc("xtick",labelsize="20"); rc("ytick",labelsize="20")
rc("axes", labelweight="bold")
```

```python
def rho_p(r, M0,a):
    return 3.0 * M0/(4.0 * np.pi * a**3) * (1.0 + r**2/a**2)**(-5.0/2.0)
```

```python
M0=1.0; a = 1.0

r_arr = np.linspace(0.0, 10.0, 1000)
rho_p_arr = np.array([rho_p(r, M0, a) for r in r_arr])
```

```python
plt.figure(figsize=(8,8))
plt.plot(r_arr, rho_p_arr, ls ='-', lw=2.5, color='black')

#plt.xscale('log')
#plt.yscale('log')

plt.xlabel(r'$r$')
plt.ylabel(r'$\rho(r)$')
#plt.savefig('plummerprofile.png')
plt.show()
```

```python
def r_str(m, a):
    return a * (m**(-2./3.) - 1.0)**(-1./2.) 
```

```python
p_res = 32
phi_arr = np.random.uniform(0.0, 2.0 * np.pi, p_res**3)
theta_arr = np.arccos( np.random.uniform(-1.0,1.0, p_res**3))
m_arr = np.random.uniform(0.0, 1.0, p_res**3)
```

```python
r_arr = np.array([r_str(m, 1.0) for m in m_arr])
x = r_arr * np.sin(theta_arr) * np.cos(phi_arr)
y = r_arr * np.sin(theta_arr) * np.sin(phi_arr)
z = r_arr * np.cos(theta_arr)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y, s=3, marker='.', color='black')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z, s=3, marker='.', color='black')
plt.show()
```
