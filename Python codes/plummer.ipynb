# Creat Plummer Sphere 
$M(r) = \frac{3}{4 \pi a^3} M_0 (1+r^2/a^2 )^{-5/2}$

```python
import numpy as np
import matplotlib.pylab as plt
from numpy.random import rand, randint
```

```python
r = rand(10)
theta = rand(10); phi = rand(10)
```

```python
x =[]; y = []; z = []
for i in range(10):
    for j in range(10):
        for k in range(10):
            x.append(r[i] * np.cos(phi[j])* np.sin(theta[k]))
            y.append(r[i] * np.sin(phi[j])* np.sin(theta[k]))
            z.append(r[i] * np.cos(theta[k]))
```

```python
print(len(x))
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.show()
```
